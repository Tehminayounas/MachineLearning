```python
class student():
    def __init__(self, name,cgpa):
        self.name = name
        self.cgpa = cgpa
        
    def get_cgpa(self):
        return self.cgpa

s1 = student('sana', 4)
s2 = student('taha',2)

print(s1.name,s1.get_cgpa())
print(s2.name,s2.get_cgpa())        
```

```python
class Animal():
  def __init__(self,name):

   def eat(self):  
      print(f"{self.name} eats food." )

class Dog(Animal):
  def bark(self):
      print(f"{self.name} barks.")
dog = Dog("Buddy")
dog.eat()
dog.bark()        
```
