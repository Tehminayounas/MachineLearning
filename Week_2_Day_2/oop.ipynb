Task 1: Employee Management System (Level 1)

```python
class Employee:
    def __init__(self, name, age, salary, department):
        self.name = name
        self.age = age
        self.salary = salary
        self.department = department
        
    def increase_salary(self, amount):
        self.salary += amount
        
        
    
    def show_details(self):
        return (
            f"Name:{self.name}\n"
            f"Age : {self.age}\n"  
            f"salary : {self.salary}\n" 
            f"Department : {self.department}\n"
        )
        
        
    
          
        
emp1 = Employee("John", 25, 10000, "social influencer")
emp2 = Employee("Taka" , 27 , 25000 , "Hr")

   
#function call print(show_details(empl1))
#method call emp1.show_details



emp1.increase_salary(5000)
emp2.increase_salary(2000)
print(emp1.show_details()) 
print(emp2.show_details()) 



       
        
        
        
        
   
```
